# Orphan Barcode/Histograms

```python
threshold=2 #dummythreshold
```

```python
def orphanbarcode(df_barcode, df_injection, threshold):
    a_strict=np.where(df_barcode>0) #pull non-zero barcodes
    b_strict=np.where(df_injection==0) #index barcodes that do not appear in OFC
    
    orphan_index=list(set(a_strict[0]) & set(b_strict[0])) #intersection non-zero barcodes that also do not appear in OFC
    remainder_index=set(a_strict[0])-set(orphan_index) #barcodes that are accounted for in OFC
    
    sns.distplot(df_barcode.loc[a_strict[0]], kde=False, label='Total Non-Zero Barcodes')
    sns.distplot(df_barcode.loc[remainder_index], kde=False, label='Non Orphan Barcodes')
    sns.distplot(df_barcode.loc[orphan_index], kde=False, label='Orphan Barcodes')
    
    plt.yscale('log')
    plt.axvline(x=threshold)
    plt.xlabel('Molecule Count')
    plt.ylabel('Barcode Count')
    plt.legend()
```

```python
def barcodecount(df_barcode, df_injection, threshold):
    a_strict=np.where(df_barcode>0) #pull non-zero barcodes
    b_strict=np.where(df_injection==0) #index barcodes that do not appear in OFC
    
    a_threshold=np.where(df_barcode>=threshold) #pull barcodes greater than or equal to threshold
    b_threshold=np.where(df_injection<threshold) #index barcodes that fall below threshold in OFC
    
    oi_strict=list(set(a_strict[0]) & set(b_strict[0]))
    oi_threshold=list(set(a_threshold[0]) & set(b_threshold[0]))
    
    remainder_index=set(a_strict[0])-set(oi_strict) #barcodes that are accounted for in OFC, nonzero
    ri_threshold=set(a_threshold[0])-set(oi_threshold) #barcodes that are accounted for in OFC, above threshold


    
    
    print('\n# Total non-zero barcodes:', len(a_strict[0]))
    print('\n# Total barcodes that do not appear in OFC:', len(b_strict[0]))
          
    print('\n# Orphan Barcodes (OFC=0, Target= non-zero):', len(oi_strict))
    print('\n# Remaining Barcodes (OFC=non-zero, Target= non-zero): ', len(remainder_index))
    print('\n% Strict Orphan Barcodes (Orphan Barcodes/Total Non-Zero Barcodes): ', (len(oi_strict)/len(a_strict[0]))*100)

    print('\n\n# Total barcodes above threshold: ', len(a_threshold[0]))
    print('\n# Total barcodes below threshold in OFC: ', len(b_threshold[0]))
          
    print('\n# Orphan Barcodes above threshold (OFC<threshold, Target>=threshold):', len(oi_threshold))
    print('\n# Remaining Barcodes above threshold: ', len(ri_threshold))
    print('\n% Threshold Orphan Barcodes (Orphan Barcodes/Total Barcodes Above Threshold): ', (len(oi_threshold)/len(a_threshold[0]))*100)

    
```

```python
#total barcodes
orphanbarcode(short_barcodemat.totalcount, short_barcodemat.OFC, 1000)
```

```python
subcortical_index=np.where((short_barcodemat.subcorticalsites>0) & (short_barcodemat.OFC!=0))
subcortical_index=set(subcortical_index[0])
    
cortical_index=np.where((short_barcodemat.corticalsites>0) & (short_barcodemat.OFC!=0))
cortical_index=set(cortical_index[0])
    
orphan_corticalindex=np.where((short_barcodemat.corticalsites>0) & (short_barcodemat.OFC==0))
orphan_corticalindex=set(orphan_corticalindex[0])
    
orphan_subcorticalindex=np.where((short_barcodemat.subcorticalsites>0) & (short_barcodemat.OFC==0))
orphan_subcorticalindex=set(orphan_subcorticalindex[0])
    
num_overlappingbarcodes=subcortical_index.intersection(cortical_index)
    
#values to plot
subcortical_frac=len(subcortical_index-num_overlappingbarcodes)
cortical_frac=len(cortical_index-num_overlappingbarcodes)
overlapping_frac=len(num_overlappingbarcodes)
subcorticalorphan_frac=len(orphan_subcorticalindex)
corticalorphan_frac=len(orphan_corticalindex)
    
sizes=[subcortical_frac, cortical_frac, overlapping_frac, subcorticalorphan_frac, corticalorphan_frac]
total=np.sum(sizes)

labels='non-zero subctx barcodes {0:.4f}'.format(subcortical_frac/total*100),'non-zero ctx barcodes {0:.4f}'.format (cortical_frac/total*100), 'overlapping barcodes in ctx/subctx {0:.4f}'.format(overlapping_frac/total*100),'orphan subctx {0:.4f}'.format(subcorticalorphan_frac/total*100), 'orphan ctx {0:.4f}'.format(corticalorphan_frac/total*100)

colors=['red','skyblue','lightseagreen','lightsalmon',
       'steelblue']
fig1=plt.pie(sizes, startangle=90, colors=colors)

plt.legend(loc = 'lower left', labels=labels)
```
